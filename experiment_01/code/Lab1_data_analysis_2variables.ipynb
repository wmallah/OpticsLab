```python
# Try to fit the data to a cos function

import scipy as sp

def cos(x, A, B):
    return A*np.cos(2*x + 2*B) + A

optimal_vals = sp.optimize.curve_fit(cos, no_beaker_polarization, no_beaker_intensity, p0=(2, 0), bounds=(0,np.pi))

x = np.linspace(np.min(no_beaker_polarization), np.max(no_beaker_polarization), 10000)

plt.scatter(no_beaker_polarization, no_beaker_intensity, label='No Beaker')
plt.plot(x, cos(x, *optimal_vals[0]), c='k', label='No Beaker Fit')
plt.xlabel('Polarization Angle (radians)')
plt.ylabel('Intensity (arb. units)')
plt.title('No Beaker')
plt.legend(loc='best')
plt.show()
print(optimal_vals[0])
```
